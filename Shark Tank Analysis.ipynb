```python

# Assuming the DataFrame is named 'df' and contains the specified columns

# Total number of episodes
total_episodes = df['Episode Number'].nunique()

# Total number of pitches
total_pitches = df.shape[0]

# Total investment amount
total_investment = df['Investment Amount (In Lakhs INR)'].sum()

# Total debt amount
total_debt = df['Debt (In lakhs INR)'].sum()


# Assuming the DataFrame is named 'df' and the columns are named as mentioned
individuals = ['Anupam', 'Ashneer', 'Namita', 'Aman', 'Peyush', 'Vineeta', 'Ghazal']

# Mapping 'Y' to 1 and 'N' to 0 for each individual
for person in individuals:
    df[person] = df[person].apply(lambda x: 1 if x == 'Y' else 0)

# Print the updated DataFrame
print(df)

# Printing the summary
print('Summary:')
print('-----------------------------------')
print('Total Episodes:', total_episodes)
print('Total Pitches:', total_pitches)
print('Total Investment:', total_investment, '(In Lakhs INR)')
print('Total Debt:', total_debt, '(In Lakhs INR)')
print('individuals', individuals)


```
